```python
#---------------------------------------Using Funtion we are Transform the job col-------------------
def f(row):
    val=row['Job']
    if val.startswith('Exec Assistant'):
        return val.removeprefix('Exec Assistant,').strip()
    elif val.startswith('Store '):
        return val.removeprefix('Store ').strip()
    elif val.startswith('Meats'):
        return val.removeprefix('Meats').strip()
    elif val.startswith('Director,'):
        return 'Director'
    elif val.endswith('Analyst'):
        return 'Analyst'
    elif val.endswith('Admin'):
        return 'Admin'
    elif val.endswith('Manager'):
        return 'Manager'
    elif val.endswith('Clerk'):
        return 'Clerk'
    else:
        return val
df['Job']=df.apply(f,axis=1)
```

```python
g_hirings=df.groupby([df['Hire_date'].dt.year,'Gender']).agg('size').reset_index(name='Hirings')
g_hirings
```

```python
f=df.loc[df['Status']=='Terminated']
```

```python
g_terminations=df.groupby([f['Term_date'].dt.year,'Gender']).agg('size').reset_index(name='Terminations')
g_terminations
```

```python
g_hirings=g_hirings.rename(columns={'Hire_date':'Year'})
g_terminations=g_terminations.rename(columns={'Term_date':'Year'})
```

```python
df2=pd.merge(g_hirings,g_terminations,how='outer').fillna(0)
df2
```

```python
df2=df2.astype({'Year':int,'Hirings':int,'Terminations':int})
```

```python
df2['Year_Wise_Retention']=df2['Hirings']-df2['Terminations']
df2['Overall']=df2['Year_Wise_Retention'].cumsum()
```
